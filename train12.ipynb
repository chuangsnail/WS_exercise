```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sklearn
import math
import joblib
from matplotlib import pyplot as plt
```

```python
# Load the npz file
train_1 = np.load("data/train_data.npy")
train_2 = np.load("data/train_labels.npy")
train_df = DataFrame(train_1)
del train_1
train_lb_df = DataFrame(train_2)
del train_2

# For validation
val_1 = np.load("data/eval_data.npy")
val_2 = np.load("data/eval_labels.npy")
val_df = DataFrame(val_1)
del val_1
val_lb_df = DataFrame(val_2)
del val_2

# for unit_test

uni_df = train_df.iloc[100:,:]
uni_lb_df = train_lb_df.iloc[100:,:]
```

```python
def LDATraining(X, y, name=""):
    from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
    from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve

    LDA = LinearDiscriminantAnalysis(n_components=1)
    LDA.fit(X, y)
    # Store the best model
    final_model = LDA
    model_name = "Model/12_final_LDA_training_" + name
    joblib.dump( final_model, model_name )
    print("Accuracy : ", accuracy_score(val_lb_df.loc[(val_lb_df[0]==1)|(val_lb_df[0]==2),:],  final_model.predict(val_df.loc[(val_lb_df[0]==1)|(val_lb_df[0]==2),:])))

    return final_model
```

```python
LDATraining(train_df.loc[(train_lb_df[0]==1)|(train_lb_df[0]==2),:],
            train_lb_df.loc[(train_lb_df[0]==1)|(train_lb_df[0]==2),0], "_new")
```
